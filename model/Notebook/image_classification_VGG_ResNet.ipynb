```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, AveragePooling2D, BatchNormalization, Add, Activation, Flatten, Dense
```

```python
image_size = (224, 224)  # Input size expected by AlexNet and VGG16
batch_size = 16

datagen = ImageDataGenerator(
    rescale=1.0/255.0,  # Normalize pixel values between 0 and 1
    validation_split=0.2  # Split the data into training and validation sets
)
```

```python
train_generator = datagen.flow_from_directory(
    parent,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    parent,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

num_classes = train_generator.num_classes
print(num_classes)
```

```python
from tensorflow.keras.applications import ResNet50

resnet_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

for layer in resnet_model.layers:
    layer.trainable = False

top_layer = tf.keras.layers.GlobalAveragePooling2D()(resnet_model.output)
top_layer = tf.keras.layers.Dense(num_classes, activation='softmax')(top_layer)

```

```python
resnet_model = tf.keras.models.Model(inputs=resnet_model.input, outputs=top_layer)
```

```python
resnet_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
resnet_history = resnet_model.fit(train_generator, epochs=12, validation_data=validation_generator)
```

```python
resnet_model.save('resnet_model.h5')
```

```python
import matplotlib.pyplot as plt

def show_plot(history,string):
    plt.plot(resnet_history.history[string])
    plt.plot(resnet_history.history["val_" + string])
    plt.xlabel("epochs")
    plt.ylabel(string)
    plt.legend([string,"val_" + string])
    plt.show()
    
show_plot(resnet_history,"accuracy")  
show_plot(resnet_history,"loss")
```

```python
resnet_history = resnet_model.fit(train_generator, initial_epoch=12, epochs=17, validation_data=validation_generator)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

model = load_model('new_vgg_model.h5')

img_path = 'ole.jpg' 
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = x/255.0
x = np.expand_dims(x, axis=0)


predictions = model.predict(x)
predicted_class_indices = np.argsort(predictions[0])[::-1][:3]

class_indices = train_generator.class_indices
class_labels = list(class_indices.keys())
predicted_labels = [class_labels[i] for i in predicted_class_indices]

print('Top three predicted classes:', predicted_labels)
```
